```python
import matplotlib.pyplot as plt
import pandas as pd
us_retail_data = pd.read_csv('C:\\Users\\aswin\\Documents\\Python AI\\Matplot lib Project\\Matplot lib Project\\us_retail_data.csv')
#print(us_retail_data.head())

#Monthly Revenue Trend
us_retail_data['Revenue'] = us_retail_data['Order_Value'] * us_retail_data['Quantity']
monthly_revenue = us_retail_data.groupby('Month')['Revenue'].sum()

#create a line plot to visualize the monthly revenue trend
plt.figure(figsize=(10, 6))

# sort months in calendar order
month_order = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]

for month in month_order:
    sorted_series = monthly_revenue.reindex(month_order)#reindexing the series to match the month order


plt.plot(sorted_series.index, sorted_series.values, marker='o', linestyle='-', color='b')
plt.title('Monthly Revenue Trend')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.grid(True)
plt.show()

#Insight: Revenue peaks in April and December, likely due to seasonal shopping trends.
```

```python

#City-wise Sales Analysis

city_sales = us_retail_data.groupby('City')['Revenue'].sum().sort_values(ascending=False)
#print(city_sales)

#create a bar chart to show city wise sales
plt.figure(figsize=(12, 6))
plt.bar(city_sales.index, city_sales.values)
plt.title('City-wise Sales')
plt.xlabel('City')
plt.ylabel('Revenue')
plt.xticks(rotation=45)# Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()

#Insight: Dallas and New York are the top-performing cities in terms of sales revenue.
```

```python
#A pie chart to display the category distribution of sales

category_sales = us_retail_data.groupby('Category')['Revenue'].sum()
plt.figure(figsize=(8, 8))
plt.pie(category_sales, labels=category_sales.index, autopct='%1.1f%%', startangle=140)#autopct to display percentage on pie chart,startangle to rotate the start angle of pie chart
plt.title('Category-wise Sales Distribution')
plt.show()

#Insight: Electronics and Kitchen categories contribute the most to sales revenue.
```

```python
#create a donut chart to show payment method distribution

payment_method_counts = us_retail_data.groupby('Payment_Mode')['Revenue'].sum()
plt.figure(figsize=(8, 8))
plt.pie(payment_method_counts, labels=payment_method_counts.index, autopct='%1.1f%%', startangle=140, wedgeprops=dict(width=0.4))
plt.title('Payment Method Distribution')
plt.legend(payment_method_counts.index, title="Payment Modes", loc="right", bbox_to_anchor=(1, 0, 0.5, 1))#loc options
plt.show()

#Insight: Amazon Pay and PayPal are the most preferred payment methods among customers.
```

```python

order_counts = us_retail_data['Customer_ID'].value_counts()
#print(order_counts)

#Create a histogram to display the customer order frequency
plt.figure(figsize=(10, 6))
plt.hist(order_counts, bins=10, color='purple', edgecolor='black')
plt.title('Customer Order Frequency')
plt.xlabel('Number of Orders')
plt.ylabel('Number of Customers')
plt.grid(True)
plt.show()

#Insight: A significant number of customers place multiple orders, indicating potential for customer retention strategies.
```
